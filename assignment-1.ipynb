```python
# Library Declarations
import numpy as np
import random

np.random.seed(0)
random.seed(0)
```

```python
def rock_paper_scissor(history=None, last_choice=None):
    # Weigh opponent and self totals with gradient weighting
    opp_totals = {}
    self_totals = {}
    total_weight = 0
    
    # Count moves with gradient weights (most recent moves get higher weight)

    for i in range(len(history)):
        position = i  # Position from oldest 
        moves_ago = len(history) - 1 - position  
        
        # Weights based on how recent moves are 
        if moves_ago < 5:
            weight = 5
        elif moves_ago < 10:
            weight = 4
        elif moves_ago < 30:
            weight = 3  
        elif moves_ago < 50:
            weight = 2 
        else:
            weight = 1  
        
        # Now apply weight
        round = history[i]
        opp = round[0]
        self = round[1]
        

        opp_totals[opp] = opp_totals.get(opp, 0) + weight
        self_totals[self] = self_totals.get(self, 0) + weight
        total_weight += weight

    opp_ratio = []
    self_ratio = []

    # Weighted ratios
    for i in [0, 1, 2]:  
        opp_ratio.append(opp_totals.get(i, 0) / total_weight)  
        self_ratio.append(self_totals.get(i, 0) / total_weight)

    opp_ratio = np.array(opp_ratio)
    self_ratio = np.array(self_ratio)

    res = []

    #rock, paper, scissor
    res.append(opp_ratio[2] + self_ratio[1])  
    res.append(opp_ratio[0] + self_ratio[2])  
    res.append(opp_ratio[1] + self_ratio[0])  
    
    res_array = np.array(res)
    if res_array.sum() == 0:
        # Fallback to uniform probabilities
        probabilities = np.array([1/3, 1/3, 1/3])
    else:
        probabilities = res_array / res_array.sum()
    
    print(probabilities)

    return np.random.choice([0, 1, 2], p=probabilities)
```

```python
# Load training data
history = np.loadtxt(r"./training.txt", dtype=int).reshape(-1,3)
print(f"Loaded training data with {history.shape[0]} rounds.")

# Use the algorithm to predict next move
print(f'Algorithm says you should choose: {rock_paper_scissor(history)}')
```
